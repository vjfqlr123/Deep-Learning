```python
# 딥러닝을 구동하는 데 필요한 케라스 함수를 불러옵니다.
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers # 케라스의 옵티마이저를 임포트
from tensorflow.keras.callbacks import EarlyStopping
```

```python
import numpy as np
import tensorflow as tf
```

```python
# 실행시 같은 결과 출력 위해 설정하는 부분
np.random.seed(3)
tf.random.set_seed(3)
```

```python
# 데이터 불러옴
Data_set = np.loadtxt('./dataset/class.csv', delimiter = ",")
Data_set
```

```python
X = [i[0] for i in Data_set]
Y = [i[1] for i in Data_set]
print(X)
print(Y)
```

```python
# tensorflow를 이용한 모델 만들기
model = Sequential()
model.add(Dense(1, input_dim = 1, activation = 'linear'))
```

```python
# 딥러닝을 실행합니다.
sgd = optimizers.SGD(lr=0.01) # 학습률 설정
model.compile(loss='mean_squared_error', optimizer=sgd, metrics=['mse'])
early_stopping = EarlyStopping()
model.fit(X, Y, epochs=800, batch_size=1, shuffle=False, callbacks=[])
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(X, model.predict(X), 'blue', X, Y, 'r.')
plt.show()
```

```python
#Gradient decent로 계산한 W와 b출력
model.layers[0].get_weights()
```

```python
b = model.layers[0].get_weights()[1][0]
b
```
