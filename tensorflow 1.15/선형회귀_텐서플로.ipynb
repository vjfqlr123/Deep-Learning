# X 와 Y의 상관관계를 분석하는 기초적인 선형회귀 모델을 만들고 실행
# 텐서 플로를 이용해 입력값(X) 이 출력값(Y)으로 변환되는 함수
# X * w + b = Y 에서 w 와 b의 값을 찾는 것이 선형회귀 모델입니다.

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

tf.enable_eager_execution()
tf.__version__
```

```python
# X는 독립변수
X = np.array([1,2,3], dtype = "float32")
# Y는 종속변수
Y = np.array([2, 2.5, 3.5], dtype = "float32")
```

```python
# X에 곱해지는 값 w를 선언, 초기값 = 2
w = tf.Variable([2], dtype = "float32")
# X에 더해지는 값 b를 선언, 초기값 = 1
b = tf.Variable([1], dtype = "float32")
```

```python
w
```

```python
b
```

```python
#learning_rate
learning_rate = 0.1
```

```python
# 300번 반복
for i in range(1000):
    # Gradient Descent를 실행해서 w와 b를 업데이트 할 객체 tape를 생성
    with tf.GradientTape() as tape:
        hypothesis = w * X + b
        
        cost = tf.reduce_mean(tf.square(hypothesis - Y))
        print(cost)
    
    w_grad, b_grad = tape.gradient(cost, [w, b])
    w.assign_sub(w_grad * learning_rate)
    
    b.assign_sub(b_grad * learning_rate)
    
```

```python
print("i={},w={},b={},cost={}".format(i, w.numpy(), b.numpy(), cost))
```
