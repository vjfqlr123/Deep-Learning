```python
import numpy as np
```

```python
# X는 독립변수
X = np.array([1,2,3], dtype = "float32")
# Y는 종속변수
Y = np.array([2, 2.5, 3.5], dtype = "float32")
```

```python
X
```

```python
Y
```

```python
# 초기 선형회귀 식 = wX + b = 2X + 1
w = 2
b = 1
```

```python
# 학습률 0 ~ 1 사이의 값
learning_rate = 0.1
```

```python
# X*w+b를 hypothesis에 대입
hypothesis = X*w+b
hypothesis
```

```python
# Gradient Decent 편미분 공식에 의해서 w_gred 계산
w_gred = 1/3*np.dot((hypothesis-Y), X.T)
w_gred
```

```python
# w 업데이트
w = w- w_gred
w
```

```python
# Greadient Decent 편미분 공식에 의해서 b_gred계산
b_gred = 1/3 * np.sum(hypothesis - Y)
b_gred
```

```python
# b 업데이트
b = b - b_gred
b
```

```python
# Gradient Decent 알고리즘을 반복 실행해서 오차제곱의 합이 최소화 되는 w와 b를 찾음
for i in range(1000):
    hypothesis = X*w+b
    print(hypothesis)
    MSE = np.sum((hypothesis-Y)**2)
    print(MSE)
    w_gred = 1/3*np.dot((hypothesis-Y), X.T)
    print(w_gred)
    b_gred = 1/3 * np.sum(hypothesis - Y)
    print(b_gred)
    w = w- (w_gred * learning_rate)
    print(w)
    b = b - (b_gred * learning_rate)
    print(b)
```
